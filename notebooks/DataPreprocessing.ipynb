```python
import pandas as pd
```

```python
# Load the CSV file into a DataFrame
df1 = pd.read_csv("nearby_cities_with_distances_under_60.csv")

# Display the first few rows
#print(df1.head())
df1 = df1['Distance (km)']
```

```python
# Load the CSV file into a DataFrame
df2 = pd.read_csv("Distance_by_time.csv")

df2['Distance (km)'] = df1.values
print(df2.head())
```

```python
df2
```

```python
# Load the CSV file into a DataFrame
df_jobs = pd.read_csv("UniForMe_Jobs.csv")
df_jobs = df_jobs.drop(df_jobs.columns[0],axis=1)
print(df_jobs.head())
```

```python
# Merge on 'From' (df2) and 'city' (df_jobs)
df_jobs['city'] = df_jobs['city'].str.lower()
df3 = df2
df3['From'] = df3['From'].str.lower()

merged_df = pd.merge(df3, df_jobs, left_on='From', right_on='city', how='left')

# Rename the 'no of jobs' column for clarity
merged_df.rename(columns={'no of jobs': 'Jobs in City'}, inplace=True)

merged_df[0:25]
```

# Universtiy Data 

```python
# Load the CSV file into a DataFrame
df_uni = pd.read_csv("daad_master_computer_science_uni.csv")
df_uni 

```

```python
value_counts = df_uni['Subject'].value_counts()
print(value_counts)
```

```python
# # Display all rows and columns
# pd.set_option('display.max_rows', None)  # None means no limit for rows
# pd.set_option('display.max_columns', None)  # None means no limit for columns

df_uni_cs  = df_uni[df_uni['Subject'] == 'Computer Science'] 
```

```python

df_uni_cs['City'] = df_uni_cs['City'].str.lower()
final_df = pd.merge(df_uni_cs, merged_df, left_on='City', right_on='From', how='inner')

final_df
```
